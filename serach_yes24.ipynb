```python
# 날짜 채우기
numsell=[]
rating = []
book_price = []
review_count = []
total_page = []
weight = []
size = []
category = []
release_date = []

for k in range(0, 999):
    h = book_href[k]
    href_url = f'https://www.yes24.com/{h}'
    href_page = requests.get(href_url)
    href_soup = BeautifulSoup(href_page.content, 'html.parser')
    sellnum = href_soup.find_all('span', class_='gd_sellNum')
    
    if sellnum == list():#판매지수
        numsell.append(None)
    else:
        sell_num = re.sub(r'[^0-9]', '', sellnum[0].get_text())
        numsell.append(sell_num)
    
    hr = href_soup.find_all('span', class_='gd_rating') # 평점
    hrc = href_soup.find_all('span', class_='gd_reviewCount')  # 리뷰수
    hp = href_soup.find_all('span', class_='nor_price') # 책 가격
    
    # 빈 리스트인 경우 0을 추가
    if hp == list():
        book_price.append(None)
    else:
        book_price.append(hp[0].get_text())
    
    # 빈 리스트인 경우 0을 추가
    if hr == list():        
        rating.append(None)
    else:
        rating.append(hr[0].get_text('|', strip=True).split('|')[1])
    
    # 빈 리스트인 경우 0을 추가
    if hrc == list() or hrc[0].find('em') == None:
        review_count.append(None)
    else:
        review_count.append(hrc[0].find('em').get_text())
        
    a = href_soup.find_all('tbody', class_='b_size')#총폐이지, 무게, 크기
    if a == list():
        total_page.append(None)
        weight.append(None)
        size.append(None)
    else:
        for word in a[0].get_text(' ',strip=True).split():
            if "쪽" in word and word != "쪽수,":
                total_page.append(word)
            if "g" in word:
                weight.append(word)
            if "mm" in word:
                size.append(word)
        if a[0].get_text(' ',strip=True).find('g') == -1:
            weight.append(None)
        if a[0].get_text(' ',strip=True).find('mm') == -1:
            size.append(None)
    
    hca = href_soup.find_all('div', class_='gd_infoSet infoSet_txtCont')
    
    if hca == list():#카테고리
        category.append(None)
    else:
        category.append(hca[0].get_text('|', strip=True).split('|')[4])
        
    hy = href_soup.find_all('tbody', class_='b_size')#발매날짜
    if hy == list():
        release_date.append(None)
    else:
        for word in hy[0].get_text('|',strip=True).split('|'):
            if "년" in word and word != '참고서 사용학년':
                release_date.append(word)
        if hy[0].get_text('|',strip=True).find('년') == -1:
            release_date.append(None)
```

```python
print(len(numsell))
print(len(rating))
print(len(book_price))
print(len(review_count))
print(len(total_page))
print(len(weight))
print(len(size))
print(len(category))
print(len(release_date))
```
